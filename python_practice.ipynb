```python
s1 = pd.Series({'Tom': 1, 'Dick': 2, 'Harry': 3})
s2 = pd.Series({'Tom': 9, 'Dick': 8, 'Mary': 7})
df = pd.DataFrame({'A': s1, 'B': s2})
df
```

```python
df = pd.concat([s1, s2], axis=1, sort=False)
df
```

```python
df = pd.DataFrame([s1, s2], index=['A','B']).T
df
```

```python
import string
import random
rows =52
cols=5
assert(1<=rows<=52) #min/max row count
df = pd.DataFrame(np.random.randn(rows, cols),
                 columns=['c'+str(i) for i in range(cols)],
                 index = list((string.ascii_uppercase + string.ascii_lowercase)[0:rows]))
df['groupable'] = [random.choice('abcde') for _ in range(rows)]
df
```
