#Plot Gas Prices 

```python
gas = pd.read_csv('gas_prices.csv')
gas.head()
```

```python
plt.plot(gas.Year, gas.USA)

plt.plot(gas.Year, gas.Canada)

plt.plot(gas['Year'], gas['South Korea'])

plt.legend()
```

```python
#gas.loc[: ,"Australia": "USA"] Fetch all countries for this range

plt.figure(figsize=(12,7))
plt.title("Gas Prices Across Countries Over Time")

plt.plot(gas['Year'], gas.loc[: ,"Australia": "USA"])

plt.legend(gas.loc[: ,"Australia": "USA"])

plt.xticks(gas['Year'][::3])
plt.xlabel("Year")
#plt.legend(gas.columns.values)

plt.show()
```

```python
plt.plot(gas.Year, gas.USA, 'b.-')

plt.plot(gas.Year, gas.Canada, 'r.-')

plt.plot(gas['Year'], gas['South Korea'], 'g.--')

plt.legend(gas[['USA','Canada', 'South Korea']])

plt.xlabel("Years")
plt.ylabel("US Dollars")
```

```python
plt.figure(figsize=(12,7))
for country in gas:
    if country != 'Year':
        plt.plot(gas['Year'], gas[country], marker = '.')

plt.legend(gas.loc[: ,"Australia": "USA"])
plt.title("Gas Prices Across Countries Over Time")
plt.xlabel("Years")
plt.ylabel("US Dollars")
```

```python
#Specific Countries

plt.figure(figsize=(12,5))

countries_to_look_at = ['Australia', 'Canada', 'Germany', 'Italy']
for country in gas:
    if country in countries_to_look_at:
        plt.plot(gas['Year'], gas[country], marker = '.')

plt.legend(gas.loc[: ,"Australia": "USA"])
plt.title("Gas Prices Across Countries Over Time")
plt.xlabel("Years")
plt.ylabel("US Dollars")
plt.show()
```

```python
#Specific Countries; Changing font

plt.figure(figsize=(12,5))

countries_to_look_at = ['Australia', 'Canada', 'Germany', 'Italy']
for country in gas:
    if country in countries_to_look_at:
        plt.plot(gas['Year'], gas[country], marker = '.')

plt.legend(gas.loc[: ,"Australia": "USA"])
plt.title("Gas Prices Across Countries Over Time", fontdict = {'fontweight':'bold', 'fontsize':15})
plt.xticks(gas['Year'][::3].tolist()+[2011])
plt.xlabel("Years")
plt.ylabel("US Dollars")
plt.savefig('gas_price_image.png',dpi = 300)
plt.show()
```

#Play with Fifa Data

```python
fifa = pd.read_csv('fifa_data.csv')
fifa.head()
```

```python
bins = [ 40, 50, 60, 70, 80, 90, 100]

plt.hist(fifa['Overall'],bins=bins)
plt.xticks(bins)
plt.xlabel("Skill Level")
plt.ylabel("Number of Players")
plt.title("Distribution of Player Skills in FIFA")

plt.show()
```

```python
#Change bin color

bins = [ 40, 50, 60, 70, 80, 90, 100]

plt.hist(fifa['Overall'],bins=bins, color='#34eba4')
plt.xticks(bins)
plt.xlabel("Skill Level")
plt.ylabel("Number of Players")
plt.title("Distribution of Player Skills in FIFA")

plt.show()
```

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[1]
```

```python
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[1]
```

```python
labels = ['Left', 'Right']

colors=['#abcdef','#aabbcc']

plt.pie([left, right], labels=labels, colors=colors, autopct='%.2f %%')

plt.show()
```

```python
fifa['Weight'] = [int(x.strip('lbs')) if type(x)==str else x for x in fifa['Weight']]
fifa.head()
```

```python
fifa['Weight']
```

```python
light = fifa.loc[fifa['Weight'] < 125].count()[1]
light_medium = fifa.loc[(fifa['Weight']>=125) & (fifa['Weight'] < 150)].count()[1]
medium = fifa.loc[(fifa['Weight'] >=150) & (fifa['Weight']<175)].count()[1]
medium_heavy = fifa.loc[(fifa['Weight'] >=175) & (fifa['Weight']<200)].count()[1]
heavy = fifa.loc[fifa['Weight'] >=200].count()[1]
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(5,5))

weights = [light, light_medium, medium, medium_heavy, heavy]
labels = ['Under 125', '125-150', '150-175', '175-200', 'Over 200']

explode = (.4, .2, 0, 0, .4)

plt.title("Weight Distribution of FIFA Players (in lbs)")
plt.pie(weights, labels=labels, autopct = '%.2f %%', pctdistance=0.8, explode=explode)


plt.show()
```
