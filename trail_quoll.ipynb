```python
#Part A
import json
import pandas as pd
import itertools

with open('exclusion\EXCLUSIONS_info_2482020.json') as f:
      data = json.load(f)

explicit = pd.read_excel("exclusion\explicit_param_2482020.xlsx")

trial_list = list()
quoll_list = list()
for dn in explicit["Managed Object FDN"].values:
    dn1 = dn.split(",")[0:3]
    dn1 = ','.join(dn1)
    #print("dn1 : ", dn1)
    for data1 in data['trial_register']['ENodeB']:
        if dn1 in data1.split("#")[-1]:
            trial_list.append(dn1)
            #print("Trail_suceed")
            continue
    #dn1 = ','.join(dn1)
    #print("dn1_after_trial : ", dn1)
    for data2 in data['quoll_data']['ENodeB'].keys():
        if dn1 == data2:
            #print("Quoll_suceed")
            quoll_list.append(dn1)
        
trial_list = list(dict.fromkeys(trial_list))
quoll_list = list(dict.fromkeys(quoll_list))

#Part B

exclusion = pd.read_excel("exclusion\Execution_Report_2482020.xlsx", sheet_name = "Exclusions")

trial_quoll_list = list()
for dn, dn1 in itertools.zip_longest(trial_list,quoll_list):
    for dne in exclusion["N/E Instance FDN"].values:
        if dn == dne or dn1 == dne:
            match = exclusion.loc[exclusion['N/E Instance FDN'] == dne, "Exclusion Due To"].iloc[0]
            if match == "Trial" or match == "Quoll":
                trial_quoll_list.append(dne)
trial_quoll_list = list(dict.fromkeys(trial_quoll_list))

#A == B
if trial_quoll_list:
    return True
else:
    return False
```

```python
def get_flattened_mo_dict(mo_dict):
    mo_data = {}
    for mo_dn, mo in mo_dict.items():
        try:
            mo_data[mo.mo_name][mo_dn] = mo
        except KeyError:
            mo_data[mo.mo_name] = {mo_dn: mo}
    return mo_data
```
